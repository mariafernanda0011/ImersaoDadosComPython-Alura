# Aula 02 [Imersão Dados com Python] Aprenda a Limpar e Preparar os Dados

```python
df.isnull()
```

```python
# Contando quantos dados nulos existem na base de dados.
df.isnull().sum()
```

```python
# Checando quais anos existem dentro da base:
df['ano_de_trabalho'].unique()
```

O resultado "nan" significa "not is a number", ele é utilizado para representar os valores nulos existentes nesse array.

```python
# Exibindo as linhas que apresentam o ano nulo:
df[df.isnull().any(axis = 1)]
```

```python
# Criando um DataFrame:

import numpy as np #Importando biblioteca para manipulação de dados numéricos.

df_salarios = pd.DataFrame({
    'nome': ["Ana", "Maria", "Junior", "Ed", "Ricardo"],
    'salario': [5000, np.nan, 4000, np.nan, 100000]
})

# Calculando a média dos salários e substitui os valores nulos:
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))

# Calculando a mediana desses salários e substitui os valores nulos:
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())
df_salarios
```

* fillna : serve para preencher a coluna 'salario_media' com os dados;
* mean() : é para calcular a média;
* median() : é para calcular a mediana;
* round2 : é para arredondar o resultado.

```python
df_temperaturas = pd.DataFrame({
    'Dia': ["Segunda-Feira", "Terça-Feira", "Quarta-Feira", "Quinta-Feira", "Sexta-Feira"],
    'Temperatura': [30, np.nan, np.nan, 28, 27]
})

# Completando os dados nullos com os valores anteriores: ffill()
df_temperaturas['preenchido_ffill'] = df_temperaturas['Temperatura'].ffill()
df_temperaturas
```

```python
df_temperaturas = pd.DataFrame({
    'Dia': ["Segunda-Feira", "Terça-Feira", "Quarta-Feira", "Quinta-Feira", "Sexta-Feira"],
    'Temperatura': [30, np.nan, np.nan, 28, 27]
})

# Completando os dados nullos com os valores posteriores: bfill()
df_temperaturas['preenchido_bfill'] = df_temperaturas['Temperatura'].bfill()
df_temperaturas
```

```python
df_cidades = pd.DataFrame({
    'nome': ["Maria", "Fernanda", "Junior", "Vitor"],
    'cidade': ["Joaíma", "Jequitinhonha", "Almenara", np.nan]
})

# Colocando um valor fixo quando o valor for nullo:
df_cidades['cidade_preenchida'] = df_cidades['cidade'].fillna('Não infomado.')
df_cidades
```

```python
# Gerando gráficos interativos:
import plotly.express as px
```

```python
# Calculando a média salarial por nível de experiência e ordenando
senioridade_media_salario = df_limpo.groupby('nivel_experiencia')['salario_em_dolar'].mean().sort_values(ascending = False).reset_index()

# Criando o gráfico interativo com Plotly
fig = px.bar(senioridade_media_salario,
             x = 'nivel_experiencia', y = 'salario_em_dolar',
             title='Média Salarial por Nível de Experiência',
             labels={'nivel_experiencia': 'Nível de Experiência ', 'salario_em_dolar': 'Média Salarial Anual (USD) '})

fig.show()
```

```python
remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie(remoto_contagem,
             names = 'tipo_trabalho', values = 'quantidade',
             title = 'Proporção dos Tipos de Trabalho',
             hole = 0.5, # Para tranformar a pizza em um gráfico de 'rosca'.
             labels={'tipo_trabalho': 'Tipo de trabalho', 'quantidade': 'Quantidade'})

fig.update_traces(textinfo='percent+label')
fig.show()
```

## Desafio da aula 03: Calcular o salário médio de cientistas de dados por país.

```python
# Filtrar o DataFrame para incluir apenas "Data Scientist"
df_cientistas_dados = df_limpo[df_limpo['cargo'] == 'Data Scientist']

salario_medio_pais = df_cientistas_dados.groupby('localizacao_da_empresa')['salario_em_dolar'].mean().round(2).sort_values(ascending = False).reset_index()
salario_medio_pais.columns = ['pais', 'salario_medio']

fig = px.bar(salario_medio_pais,
             x = 'pais', y = 'salario_medio',
             title = 'Salário médio de cientistas de dados por país',
             labels = {'pais': 'País', 'salario_medio': 'Salário médio (USD)'})
fig.show()
```

```python
import pycountry

# Função para converter ISO-2 para ISO-3
def iso2_to_iso3(code):
    try:
        return pycountry.countries.get(alpha_2=code).alpha_3
    except:
        return None

# Criar nova coluna com código ISO-3
df_limpo['residencia_iso3'] = df_limpo['localizacao_da_empresa'].apply(iso2_to_iso3)

# Calcular média salarial por país (ISO-3)
df_ds = df_limpo[df_limpo['cargo'] == 'Data Scientist']
media_ds_pais = df_ds.groupby('residencia_iso3')['salario_em_dolar'].mean().reset_index()

# Gerar o mapa
fig = px.choropleth(media_ds_pais,
                    locations='residencia_iso3',
                    color='salario_em_dolar',
                    color_continuous_scale='rdylgn',
                    title='Salário médio de Cientista de Dados por país',
                    labels={'salario_em_dolar': 'Salário médio (USD)', 'residencia_iso3': 'País'})

fig.show()
```
